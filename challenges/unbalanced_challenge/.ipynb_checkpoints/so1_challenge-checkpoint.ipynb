# Quick exploration

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn import cross_validation

sns.set(style="ticks")
np.random.seed(10)
```

```python
data = pd.read_csv('data/train.tsv', delimiter='\t')
```

There are 3 types of columns in the datasets:
    - binary (0 or 1, 30 columns and 367 unique row combinations)
    - float values
    - cId (from 1 to 50)

```python
def get_binary_columns(dataset):
    """Get names of binary columns"""
    dtypes = dataset.dtypes
    # to skip id, y and cId use [3: ]
    return dtypes[dtypes == np.int64].index[3: ].tolist()

def get_float_columns(dataset):
    """Get names of float columns"""
    dtypes = dataset.dtypes
    return dtypes[dtypes != np.int64].index.tolist()

bin_columns = get_binary_columns(data)
float_columns = get_float_columns(data)
```

Following report based on assumption that all cId groups are rather similar and model that works for one cId group also works for others.

# Classification

The problem is definitely not linear: 
- float features for both classes are normaly distributed around 0
- binary features are not correlated with y and between themselves

```python
from matplotlib.colors import ListedColormap

def evaluate_clf(X, y, clf, cv=5):
    """Evaluate classifier by calculating f1 score cv times"""
    
    print 'Classifier', clf
    
    if hasattr(X, 'values'):
        X = X.values
    y = y.values
    kf = cross_validation.KFold(X.shape[0], n_folds=cv, shuffle=True,
                                random_state=10)
    hist = []
    for train, test in kf:
        X_train, X_test = X[train, :], X[test, :]
        y_train, y_test = y[train], y[test]
    
        clf.fit(X_train, y_train)
        pred = clf.predict(X_test)
        f1 = metrics.f1_score(y_test, pred)
        print 'current F1:', f1
        hist.append(f1)
        
    print 'mean:', np.mean(hist)
    
    classification_report(X, y, clf)
    
def classification_report(X, y, clf):
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33,
                                                        random_state=10)
    clf.fit(x_train, y_train)
    pred = clf.predict(x_test)
    
    print 'Classification report \n'
    print metrics.classification_report(y_test, pred)
    
def plot_decision_boundaries(X, y, clf):
    """Plot 2D decision boundaries"""
    
    assert X.shape[1] == 2
    
    figure(figsize(10, 7))

    if hasattr(X, 'values'):
        X = X.values
    y = y.values
    
    h = 0.005
    x_min, x_max = X[:, 0].min() - 0.025, X[:, 0].max() + 0.025
    y_min, y_max = X[:, 1].min() - 0.025, X[:, 1].max() + 0.025
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    clf.fit(X, y)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=.8)

    idx = y == 0
    plt.scatter(X[idx][:, 0], X[idx][:, 1], alpha=0.4)
    
    idx = y == 1
    plt.scatter(X[idx][:, 0], X[idx][:, 1], alpha=0.4, color='r')
```

Visually class 0 and class 1 have sighly different distributions so we can try to fit Gaussian mixture model

```python
from sklearn.mixture import GMM

clf = GMM(n_components=2, random_state=10)
evaluate_clf(data_cid[float_columns], data_cid.y, clf)
```

```python
# transform data to 2D using PCA in order to plot decision boundaries
pca = PCA(n_components=2, whiten=True)
data_pca = pca.fit_transform(data_cid[float_columns])

figure(figsize(8, 5))
plot_decision_boundaries(data_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries for Gaussian Mixture Model')
```

And for binary data the simples we can whould be to use Gaussian Naive Bayes classifier

# Final solution

```python
# Let's plot 2D scatter plot again

figure(figsize(8, 5))
idx = (data_cid.y == 0).values
plt.scatter(data_pca[idx][:, 0], data_pca[idx][:, 1], alpha=0.4)

idx = (data_cid.y == 1).values
plt.scatter(data_pca[idx][:, 0], data_pca[idx][:, 1], alpha=0.4, color='g')
plt.legend(['Class 0', 'Class 1'])
_ = plt.title('2d scatter plot of the float featues')
```

This data doesn't look separatable in 2D space. But still, there is a chance, that is in 3D space or in any higher dimentional space it's possible to find hyperplane that splits the data. Very simple explanation of this idea: in 3D space green points can be a bit closer to the observer than in blue ones. So in 2D at least 3-rd dimention is needed so it's possible to compute polynomial features. But we have 70 dimentions (70 float features) + 30 binary dimentions. In this case we can use kernel methods like rbf or polynomial kernels.

```python
from sklearn.decomposition import KernelPCA

# pca = KernelPCA(n_components=2, kernel='rbf', gamma=0.01, alpha=0.9, fit_inverse_transform=True)
pca = KernelPCA(n_components=2, kernel='rbf', fit_inverse_transform=True)
X_pca = pca.fit_transform(data_cid[float_columns])

figure(figsize(8, 5))
_ = sns.kdeplot(X_pca, shade=True, shade_lowest=False)
```

Now there are 2 hotspots and the data is not centered around 0 anymore.

```python
figure(figsize(8, 5))

idx = (data_cid.y == 0).values
plt.scatter(X_pca[idx][:, 0], X_pca[idx][:, 1], alpha=0.4)

idx = (data_cid.y == 1).values
plt.scatter(X_pca[idx][:, 0], X_pca[idx][:, 1], alpha=0.4, color='g')
plt.legend(['Class 0', 'Class 1'])
_ = plt.title('2 componets kernel PCA with RDF kernel')
```

Now there is a certain structure compared to the initial 2D scatter plot after fitting PCA.

```python
# check the same GMM algorithm
clf = GMM(n_components=2, random_state=10)
evaluate_clf(X_pca, data_cid.y, clf)
```

```python
figure(figsize(8, 5))

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries for Gaussian Mixture Model after fitting kernel PCA')
```

But more logically would be to use classification algorithms that support kernel method, for example support vector machine

```python
from sklearn import svm

# since the number of classes is not balanced use class_weight='auto'
clf = svm.SVC(kernel="rbf", class_weight='auto', gamma=0.01,
              C=100, random_state=10)

evaluate_clf(X_pca, data_cid.y, clf)
```

```python
figure(figsize(8, 5))

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries of SVN with RBF kernel')
```

The next logical step is to combine binary and float features

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import KernelPCA

pca = KernelPCA(n_components=2, kernel='rbf', fit_inverse_transform=True)

scaler = StandardScaler()
X = scaler.fit_transform(data_cid[float_columns + bin_columns])
X_pca = pca.fit_transform(X)

figure(figsize(8, 5))
_ = sns.kdeplot(X_pca, shade=True, shade_lowest=False)
```

```python
clf = svm.SVC(kernel="rbf", class_weight='auto', gamma=0.01,
              C=100, random_state=10)

evaluate_clf(X_pca, data_cid.y, clf)
```

```python
figure(figsize(8, 5))

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries of SVN with RBF kernel')
```

```python
# finally evaluate SVC with RBF kernel on raw data without Kernel PCA
# it should perform even better

scaler = StandardScaler()
X = scaler.fit_transform(data_cid[float_columns + bin_columns])

clf = svm.SVC(kernel="rbf", class_weight='auto', random_state=10)

%time evaluate_clf(X, data_cid.y, clf)
```

```python
clf = GaussianNB()

evaluate_clf(X, data_cid.y, clf)

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries of SVN with RBF kernel')
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=2, weights='distance')
evaluate_clf(X_pca, data_cid.y, clf)

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries of KNeighborsClassifier with RBF kernel')
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=200, learning_rate=0.9,
                                 random_state=10)

evaluate_clf(X, data_cid.y, clf)

plot_decision_boundaries(X_pca, data_cid.y, clf)
_ = plt.title('Decision boundaries of GradientBoostingClassifier with RBF kernel')
```

# Solution speed optimization

The performance of SVC with RDF or KernelPCA with other classifier is damn slow. As a trick to optimize this we can use kernel approximations and LinearSVC

```python
from sklearn.svm import LinearSVC
from sklearn.kernel_approximation import RBFSampler
from sklearn.pipeline import Pipeline

kernel = RBFSampler(n_components=300, gamma=.001, random_state=10)
clf = LinearSVC(C=100., class_weight='auto')
clf = Pipeline([('rbf', kernel), ('svc', clf)])

%time evaluate_clf(data_cid[float_columns + bin_columns], data_cid.y, clf)
```
